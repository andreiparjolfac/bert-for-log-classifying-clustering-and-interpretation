```python
import pandas as pd

g4_df = pd.read_excel("data/datag4.xlsx")
g4_df = g4_df[["Script Name","Issue Type","Juice Session IDs","JuicePath.1","JuicePath.2","JuicePath.3","JuicePath.4"]]
g4_df = g4_df[g4_df["Issue Type"]!="Unknown"]
g4_df["Juice Session IDs"] = g4_df["Juice Session IDs"].apply(lambda x:len(str(x).split(" ")))

g5_df = pd.read_excel("data/datag5.xlsx")
g5_df = g5_df[["Script Name","Issue Type","Juice Session IDs","JuicePath.1","JuicePath.2","JuicePath.3","JuicePath.4"]]
g5_df = g5_df[g5_df["Issue Type"]!="Unknown"]
g5_df["Juice Session IDs"] = g5_df["Juice Session IDs"].apply(lambda x:len(str(x).split(" ")))

df=pd.concat([g4_df,g5_df])
df.head()


```

```python
from pathlib import Path
import glob
import os 

def get_pure_logs(dir_path):
    if isinstance(dir_path,str) and dir_path !="NaN":
        juice_dir = os.path.abspath(dir_path[5:])
        try:
            log = juice_dir + "/" + glob.glob("ScriptOutput_Info_*.txt",root_dir=juice_dir)[0]
            print(log)
            with open(log,encoding="utf-8") as fp:
                lines = list(fp.readlines())
                lines = " ".join(lines[-30:])
            return lines
        except Exception as e:
            return 
    else:
        return
```

```python
col = "JuicePath."
for  i in range(1,5):
    col+=str(i)
    df[col]=df[col].apply(get_pure_logs)
    col = col[:-1]
df.head()
```

```python
import re
import string
df = df.dropna(axis=0,how='all' ,subset=["JuicePath.1","JuicePath.2","JuicePath.3","JuicePath.4"])
def clean_text(text,s_name):
    if text is not None:
        text = str(text).lower()
        text = re.sub('\[.*?\]',' ',text)
        text = re.sub('https?://\S+|www\.\S+',' ',text)
        text = re.sub('<.*?>+',' ',text)
        text = re.sub('[%s]'%re.escape(string.punctuation),' ',text)
        text = re.sub('\n',' ',text)
        text = re.sub('\w*\d\w*',' ',text)
        return s_name+text
    else:
        return text

df["JuicePath.1"] = df.apply(lambda x:clean_text(x["JuicePath.1"],x["Script Name"]),axis=1)
df["JuicePath.2"] = df.apply(lambda x:clean_text(x["JuicePath.2"],x["Script Name"]),axis=1)
df["JuicePath.3"] = df.apply(lambda x:clean_text(x["JuicePath.3"],x["Script Name"]),axis=1)
df["JuicePath.4"] = df.apply(lambda x:clean_text(x["JuicePath.4"],x["Script Name"]),axis=1)

```

```python
df.to_csv("data/dataprocesed.csv")
```
